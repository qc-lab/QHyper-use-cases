# Advantage solver

## With one_hot_encoding - simple problem

```python
from QHyper.problems.community_detection import Network, CommunityDetectionProblem
import networkx as nx

G = nx.Graph()
G.add_edges_from([(0,1),(1,2),(0,2), (3,4),(4,5),(5,3), (5,0)])

G = Network(G)

problem = CommunityDetectionProblem(G, communities=2, one_hot_encoding=True)
```

```python
from QHyper.solvers.advantage import Advantage

advantage = Advantage(problem)
response_simple_with = advantage.solve()
```

```python
print(response_simple_with)
```

```python
print('-' * 60)
print('{:>15s}{:>15s}{:^15s}'.format('Comm 0','Comm 1','Energy'))
print('-' * 60)
for sample, E in response_simple_with.data(fields=['sample','energy']):
    S0 = [k for k,v in sample.items() if v == 0]
    S1 = [k for k,v in sample.items() if v == 1]
    print('{:>15s}{:>15s}{:^15s}'.format(str(S0),str(S1),str(E)))
```

```python
from demo.network_community_detection.utils.utils import draw_communities


sample = response_simple_with.first

folder = "./network_community_detection/demo_output"
data_name = "simple_problem_with_one_hot"
path = f"{folder}/{data_name}.png"

draw_communities(problem, sample.sample, path)
```
